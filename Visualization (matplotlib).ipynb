```python
#plt.subplots creates 2 different objects:- Figure object and axes object
#The figure object is a container that holds everything that you see on the page.
#Meanwhile, the axes is the part of the page that holds the data., it is the canvas on which we will draw with our data, to visualise it.
fig,ax=plt.subplots()
plt.show()
```

```python
dataset={'ID':[1,2,3,4],'Month':['Jan','Feb','Mar','Apr'],'Amount':[100,200,300,400]}
import pandas as pd
```

```python
dataframe1
```

```python
plt.plot(dataframe1['Month'],dataframe1['Amount'])
plt.plot(dataframe1['Month'],dataframe1['ID'])
plt.show()
#This one takes the current figure and axes (if none exists it will create a new one) and plot into them.
```

```python
#Adding the marker and linestyle arguments
plt.plot(dataframe1['Month'],dataframe1['ID'],marker='v',linestyle='--')
```
